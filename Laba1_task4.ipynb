```python
# -*- coding: utf-8 -*-
def form_dict():
    d = {}
    iter = 0
    for i in range(97,123):
        d[iter] = chr(i)
        iter = iter +1
    d[26] = ' ' # надо закоментить для поиска в строке из доашней работы
    return d
    

def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k
        
        
alphabet = form_dict()
```

```python
def letter_to_num(string_list):
    nums = []
    for i in string_list:
        nums.append(get_key(alphabet, i))
    return nums
```

```python
def encript(s, key):
    state = letter_to_num(s)
    key = letter_to_num(key)
    counter = 0
    while len(state) % len(key) != 0:
        state.append(0)
        counter += 1
    for i in range(len(key)):
        for j in range(0 + i,len(state), len(key)):
            state[j] = (state[j] + key[i]) % len(alphabet)
    for i in range(counter):
        state.pop()
    s1 =''
    for i in state:
        s1 += alphabet[i]
    return s1
```

```python
def decript(s, key):
    state = letter_to_num(s)
    key = letter_to_num(key)
    counter = 0
    while len(state) % len(key) != 0:
        state.append(0)
        counter += 1
    for i in range(len(key)):
        for j in range(0 + i,len(state), len(key)):
            state[j] = (state[j] - key[i]) % len(alphabet)
    for i in range(counter):
        state.pop()
    s1 =''
    for i in state:
        s1 += alphabet[i]
    return s1
```

```python
s = public_text
key = 'jan'
print(s)
print()
secret = encript(s, key)
print(secret)
print()
print(decript(secret, key))
```
